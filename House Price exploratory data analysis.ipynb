### Importing libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

pd.pandas.set_option('display.max_columns', None)
```

```python
#Loading data
dataset = pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

### Checking for missing values

```python
features_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1]

for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(), 2), "% missing values")
```

```python
#Visualizing distribution of missing values with a heatmap
plt.figure(figsize=(20,20))
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

### Exploratory data analysis

```python
# plotting the relationship of missing values with house price using a countplot
for feature in features_with_nan:
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
#Handling numerical features
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print('The number of numerical features is: {}'.format(len(numerical_features)))

dataset[numerical_features].head()
```

```python
#Handling year features
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

print('The number of year features is: {}'.format(len(year_features)))

dataset[year_features].head()
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.grid()
plt.ylabel('House Price')
plt.title('House price by year sold')
plt.show()
```

```python
#The relationship between year sold and house price
for feature in year_features:
    if feature != 'YrSold':
        data = dataset.copy()
        
    data[feature] = data['YrSold'] - data[feature]
    
    plt.scatter(data[feature], data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
#Handling discrete features
discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_features + ['Id']]

print('The number of discrete feature is: {}'.format(len(discrete_features)))
```

```python
discrete_features
```

```python
#The relationship between discrete features and house price
for feature in discrete_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('House Price')
    plt.title(feature)
    plt.show()
```

```python
#Handling continous features
continous_features = [feature for feature in numerical_features if feature not in discrete_features and feature not in year_features + ['Id']]

print('Continous feature count: {}'.format(len(continous_features)))
```

```python
#Plotting the distribution of continous features
for feature in continous_features:
    data = dataset.copy()
    data[feature].hist(bins=40)
    plt.xlabel(feature)
    plt.ylabel('House Price')
    plt.title(feature)
    plt.show()
```

```python
# Normalizing the distribution with log normal distribution and plotting the relationship with sales price
for feature in continous_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('House Price')
        plt.title(feature)
        plt.show()
```

```python
#Checking for outliers in continous features
for feature in continous_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
#Handling categorical features
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']

print('The number of categorical features is: {}'.format(len(categorical_features)))
```

```python
dataset[categorical_features].head()
```

```python
#Checking the cardinality of the categorical features 
for feature in categorical_features:
    print('The feature is {} and number of unique categories are {}'.format(feature, len(dataset[feature].unique())))
```

```python
#Plotting the relationship between categorical features and house price
for feature in categorical_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('House Price')
    plt.title(feature)
    plt.show()
```
